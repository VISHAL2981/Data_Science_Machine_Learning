# correlation with heatmap

 a correlation heatmap uses colored cells typically in  monochromatic scale to show a 2d correlation matrix(table) between two discrete dimensions or event types.it is very important in feature slection

```python
df.corr()#only take integer values
```

```python
sns.heatmap(df.corr())
```

# join plot
a join plot allows to study the relationship btw 2 numeric variables the central chart display their correlation it is usually a scatterplot a hexbin plot,a 2 d histogram or a 2d density plot



# univariate analysis

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='reg')
```

# pair plot
a pair plot is also kmown as scatterplot in which one variable in same data row is matched with with another variable value like this:pair plots are just elaborations on this showing all variables paired with all the other variables


```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
sns.pairplot(df,hue='size')
```

```python
sns.pairplot(df,hue='smoker')
```

```python
df['smoker'].value_counts()
```

# Dist plot
dist plot helps us to check the distribution of the column feature

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=False,bins=10)#kde=kernel density estimation
```

```python
sns.distplot(df['tip'],kde=False,bins=100)
```

```python
sns.distplot(df['tip'],kde=True,bins=10)
```

# categorical plots
seaborn also helps us in doing the analysis on categorial data points in this section we will discuss about 
-boxplot
-violinpllot
-countplot
-bar plot

```python
#count plot
sns.countplot('sex',data=df)
```

```python
sns.countplot('day',data=df)
```

```python
sns.countplot(y='day',data=df)
```

```python
##bar plot
sns.barplot(x='total_bill',y='sex',data=df)
```

```python
sns.barplot(x='total_bill',y='smoker',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

# box plot

a box and whisker plot is a graph represent information from a five number summary


```python
sns.boxplot('sex',"total_bill",data=df)
```

```python
sns.boxplot(x='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')
```

```python
  # categorize my data based on some other categories
sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

# violin plot
violin plot helps us to see both the distribution of data in terms of kernel density estimation and the box plot


```python
sns.violinplot(x='total_bill',y='day',data=df,palette='rainbow')
```
