# matplot Library tutorial
matplotLib is a plotting library for the python programming language and its numerical mathematic extension numpy. it provide an object oriented API for embedding plots into application using general purpose GUI toolkits like TKINTER,wxPython,Qt or GTK+

SOME OF THE MAJOR PROS OF MATPLOTLIB ARE:

- genrally easy to get started for simple plots
- support for custom labels and text
- great control of every element in a figure
- high quality output in many formats
- very customizable in genral


```python
import matplotlib.pyplot as plt
    
%matplotlib inline 
```

```python
import numpy as np
```

```python
##simple examples
x=np.arange(0,10)
y=np.arange(11,21)
```

```python
##plotting using matplot lib

## plt scatter

plt.scatter(x,y,c='r')# r is colour or c whatevr it is
## for title for x line and y line
plt.xlabel('X AXIS')
plt.ylabel('Y AXIS')
plt.title("Graph in 2D")
## we can save image also 
plt.savefig('test.png')
```

```python
y=x*x
```

```python
##plt plot
plt.plot(x,y)
```

```python
plt.plot(x,y,'r--')
```

```python
plt.plot(x,y,'ro--')
```

```python
plt.plot(x,y,'r*',linestyle='dashed',linewidth=4,markersize=16)
plt.xlabel('X axix')
plt.ylabel('Y axix')
plt.title("Graph in 2D")
```

```python
## creating subplots
plt.subplot(2,2,1)
plt.plot(x,y,'r--')
plt.subplot(2,2,2)
plt.plot(x,y,'g*--')
plt.subplot(2,2,3)
plt.plot(x,y,'bo')
plt.subplot(2,2,4)
plt.plot(x,y,'yo--')
```

```python
x=np.arange(1,11)
y=3*x+5
plt.title("Matplotlib demo")
plt.xlabel('x axis caption')
plt.ylabel("y axis caption")
plt.plot(x,y)
plt.show()
```

```python
#compute the x and y coordinated for points on a sine curve
x=np.arange(0,4*np.pi,0.1)
y=np.sin(x)
plt.title("sine wave form")

# plot the points using matplotlib
plt.plot(x,y)
plt.show()
```

```python
#subplot()
# compute the x and y coordinated for points on sine and cosine curves
x=np.arange(0,5*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)

#set up a subplot grid that has height 2 and width 1,and set the first such subplot as active.
plt.subplot(2,1,1)#it indicates (2,1,2) first two is row ,1 is column,1 is position

#make the first plot
plt.plot(x,y_sin,'r--')
plt.title('sine')

# set the second subplot as active and make the second plot
plt.subplot(2,1,2)#it indicates (2,1,2) first two is row ,1 is column,2 is position
plt.plot(x,y_cos,'g--')
plt.title('cosine')

#show the figure
plt.show()
```

```python
#bar plot

x=[2,8,12]
y=[11,16,9]

x2=[3,9,11]
y2=[6,15,11]

plt.bar(x,y)
plt.bar(x2,y2,color='g')
plt.title('bar grapgh')
plt.ylabel('y axis')
plt.xlabel('x axis')

plt.show()
```

# box plot using matplotlib

```python
data=[np.random.normal(0,std,100)for std in range(1,4)]

#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=True);
```

```python
data=[np.random.normal(0,std,100)for std in range(1,4)]

#rectangular box plot
plt.boxplot(data,vert=False,patch_artist=False);
```

# piechart

```python
 #Data to Plot
labels='python','C++','Ruby','Java'
sizes=[215,130,245,210]#we calculate percentage here by total of all which is 800 and then then we fin d percentage of one 215/800*100=26.9  like this we find all of these

colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.2,0,0,0)# distance between slices we use this

#explode 1st slice
    
#plot
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
    
plt.axis('equal')
plt.show()
```
